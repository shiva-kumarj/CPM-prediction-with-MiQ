### Bivariate analysis

#### Which browser has the most impressions overall

```python
browser_impressions = df.groupby('Browser', as_index=False)\
.sum()[['Browser', 'Impressions']]\
.sort_values(by=['Impressions'],
             ascending = False).head()
```

```python
import plotly.express as px
plt.figure(figsize = (15, 8))
fig = sns.barplot(data = browser_impressions, x = 'Browser', y = 'Impressions', )
plt.show()
```

#### Which Device Type gets the most number of impressions

```python
device_impressions = df.groupby('Device Type', as_index=False)\
.sum()[['Device Type','Impressions']]\
.sort_values('Impressions',ascending=False)

plt.figure(figsize = (15, 8))
fig = sns.barplot(data = device_impressions, x = 'Device Type', y = 'Impressions')
plt.show()
```

#### Which advertiser spent the most for impressions

```python
impression_expenditure = df.groupby("IO_ID", as_index = False)\
.sum()[['IO_ID','CPM']]\
.sort_values('CPM', ascending = False)

plt.figure(figsize = (15, 8))
sns.barplot(data = impression_expenditure, 
            x = 'IO_ID', y = 'CPM', 
            order = impression_expenditure['IO_ID'])
plt.show()
```

#### Who got the most impressions

```python
impression_bought = df.groupby("IO_ID", as_index = False)\
.sum()[['IO_ID', 'Impressions']]\
.sort_values('Impressions', ascending = False)

plt.figure(figsize = (15, 8))
sns.barplot(x = 'IO_ID', y ='Impressions', data = impression_bought, order = impression_bought['IO_ID'])
plt.show()
```

#### Which creative size is most advantageous

```python
creative_size = df.groupby('Creative Size',as_index=False)\
.sum()[['Creative Size','Impressions']]\
.sort_values('Impressions', ascending = False)

plt.figure(figsize = (15, 8))
sns.barplot(data = creative_size, x='Creative Size', y='Impressions')
plt.show()
```

#### Breakdown creative size by device type.

```python
creative_size_popularity = df.groupby(['Creative Size', 'Device Type'],\
           as_index = False).count()[['Creative Size', 
                                      'Device Type',
                                      'Impressions']].sort_values('Impressions', ascending = False)

plt.figure(figsize = (15, 8))
sns.barplot(data = creative_size_popularity, 
            x = 'Creative Size', 
            y = 'Impressions', 
            hue = 'Device Type',
           capsize=0.2)
plt.show()
```

#### Which App/URL is most expesive to get an impression in and whats the creative size of it

```python
expensive_apps = df.groupby(['App/URL ID', 'Creative Size'], as_index = False)\
.sum()[['App/URL ID', 'Creative Size','CPM']].sort_values('CPM',ascending = False).head()

plt.figure(figsize=(10, 9))
sns.barplot(x = 'App/URL ID', y = 'CPM', 
            hue = 'Creative Size', 
            data = expensive_apps,
           order = expensive_apps['App/URL ID'])
```

### Multi-Variate analysis

#### Whats the daily impression per Device Type

```python
# get day of week
import datetime as dt
df['day_of_week'] = df['Date'].dt.day_name()
```

```python
daily_impressions = df.groupby(['Date', 'Device Type'], as_index = False).\
sum()[['Date', 'Device Type', 'Impressions']].\
sort_values('Impressions', ascending = False)

plt.figure(figsize = (10, 8))
sns.lineplot(x = 'Date', y = 'Impressions',
            hue = 'Device Type', 
            data = daily_impressions)
```

#### How does CPM look like over the week

```python
from pandas.api.types import CategoricalDtype
cats = [ 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
cat_type = CategoricalDtype(categories=cats, ordered=True)

daily_sales = df.groupby(['day_of_week'], as_index = False)\
.mean()[['day_of_week','CPM']]

daily_sales['day_of_week'] = daily_sales['day_of_week'].astype(cat_type)
```

```python
# CPM over the week trend
plt.figure(figsize = (10, 8))
sns.lineplot(x = 'day_of_week', y = 'CPM', data = daily_sales)
plt.show()
```

#### Daily Distribution of CPM

```python
# The distribution of CPM looks similar to Log-normal distribution
# Applying log transformation for a better statistical analysis

df['CPM_log'] = np.log(df['CPM'])
```

```python
plt.figure(figsize = (10, 8))
sns.boxplot(y = 'CPM_log', x = 'day_of_week', data = df, palette='Set3')
plt.show()
```

#### Distribution of CPM between Devices

```python
plt.figure(figsize = (10, 8))
sns.boxplot(x = 'Device Type', y = 'CPM_log', data = df, palette = 'Set3')
plt.show()
```

##### The mean CPM for Smartphones and tablets is higher, because they are prime real-estate for ads.

#### CPM distribution for each ad campaign

```python
plt.figure(figsize=(10, 8))
sns.boxplot(y = 'CPM_log', x = 'IO_ID', data = df, palette='Set3', )
plt.show()
```

- Campaign with IO_ID 2 must have been expensive, its lowest CPM is higher than the rest

#### Distribution of CPM among the various browsers

```python
# Sort them my median, due to large number of outliers
CPM_median_browser = df.groupby('Browser', as_index = False)\
.median()[['Browser', 'CPM_log']]\
.sort_values('CPM_log', ascending = False)
```

```python
plt.figure(figsize=(10, 8))
sns.boxplot(x = 'Browser', y = 'CPM_log',
            data = df, palette='Set3',
            order = CPM_median_browser['Browser'])
plt.xticks(rotation = 90)
plt.tight_layout()
plt.show()
```

## Statistical Analysis of continuous target variable (CPM)

```python
df['CPM'].describe()
```

### Probability Density function of CPM

```python
from scipy.stats.kde import gaussian_kde
kde = gaussian_kde(df['CPM'])
dist_space = np.linspace(min(df['CPM']), max(df['CPM']), 100)
plt.figure(figsize=(10, 8))
plt.plot(dist_space, kde(dist_space))
```

### CDF of the continuous variable: This looks like log-normal distribution

```python
from scipy import stats
plt.figure(figsize=(10, 8))
stats.probplot(df['CPM'], plot = plt)
plt.show()
```

### Applying log transformation for a better statistical analysis

```python
# PDF of log(CPM)
from scipy.stats.kde import gaussian_kde
kde = gaussian_kde(np.log(df['CPM']))
dist_space = np.linspace(min(np.log(df['CPM'])), max(np.log(df['CPM'])), 100)
plt.figure(figsize=(10, 8))
plt.plot(dist_space, kde(dist_space))
```

### CDF of log(CPM)

```python
from scipy import stats
plt.figure(figsize = (10, 8))
stats.probplot(np.log(df['CPM']), plot = plt)
plt.show()
```

```python
# Viz the distribution using boxplot
plt.figure(figsize=(10, 8))
sns.boxplot(y = np.log(df['CPM']))
plt.show()
```

## Time-Series EDA

### Trend of mean browser cpm

```python
# top 5 broswer CPMs
browser_cpm = df.groupby('browser', as_index = False)\
.mean()[['browser', 'cpm']].sort_values('cpm', ascending = False)
browser_cpm.head()
```

#### Trend of median CPM for top 5 browsers

```python
%matplotlib inline
import matplotlib.pyplot as plt
top_browser_list = list(browser_cpm['browser'].iloc[1:6])

browser_cpm_trend = pd.DataFrame(df[(df['browser'].isin(top_browser_list))])[['date', 'browser', 'cpm']]\
.groupby(['date', 'browser'])['browser', 'cpm'].median().unstack()

browser_cpm_trend.plot(figsize = (15, 8), linewidth = 2, grid = True,
                      title = 'Trend of Mean CPM of Browsers: lower is better')
plt.show()
```

    The trend looks erratic but in this timeframe there is a reduction in 
    the mean CPM towards the end.
    - mean CPM for the famous browsers like Chrome, Safari dropped towards the end.

### Trend of the mean CPMs per creative size

```python
# top 5 broswer CPMs
creative_cpm = df.groupby('creative_size', as_index = False)\
.mean()[['creative_size', 'cpm']].sort_values('cpm', ascending = False)

creative_cpm.head()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
creative_cpm_list = list(creative_cpm['creative_size'].iloc[1:6])

browser_cpm_trend = pd.DataFrame(df[(df['creative_size'].isin(creative_cpm_list))])\
[['date', 'creative_size', 'cpm']]\
.groupby(['date', 'creative_size'])['creative_size', 'cpm'].median().unstack()

browser_cpm_trend.plot(figsize = (15, 8), linewidth = 2, grid = True,
                      title = 'Trend of Median CPM for creative size: lower is better')
plt.show()
```

    The median CPM of most creative size remains stable over this time-frame.
